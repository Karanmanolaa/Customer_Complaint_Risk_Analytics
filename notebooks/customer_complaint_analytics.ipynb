```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#saving the dataset in the variable df
df = pd.read_csv("/content/complaints-2026-07-02_13_47.csv")
```

```python
#lets see the first 5 values of the dataset
df.head()
```

```python
#what's the shape of the dataset?
df.shape
```

```python
#Name of all the columns
df.columns.tolist()
```

```python
#fixing the column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-","_")
    .str.replace("?", "", regex=False)
)

df.columns.tolist()
```

```python
#lets review the shape and view the first 5 values
print(df.shape)
display(df.head())
```

```python
# Convert date columns to datetime format
df["date_received"] = pd.to_datetime(df["date_received"], errors="coerce")
df["date_sent_to_company"] = pd.to_datetime(df["date_sent_to_company"], errors="coerce")

df[["date_received", "date_sent_to_company"]].head()
```

```python
# Check the date range and total number of rows
print("Earliest date:", df["date_received"].min())
print("Latest date:", df["date_received"].max())
print("Rows:", len(df))
```

```python
#missing values
missing_count = df.isna().sum()
missing_count
```

```python
# Calculate the percentage of missing values in each column and then displaying them
missing_percent = (df.isna().sum() / len(df) * 100).round(2)
missing_percent.sort_values(ascending=False)
```

```python
# Replace missing text values with clear labels
df["tags"] = df["tags"].fillna("No tag")
df["company_public_response"] = df["company_public_response"].fillna("No public response")
```

```python
# Replaced missing subissues with a clear label
df["sub_issue"] = df["sub_issue"].fillna("Not specified")
```

```python
# Replace missing location details with "Unknown"
df["state"] = df["state"].fillna("Unknown")
df["zip_code"] = df["zip_code"].fillna("Unknown")
```

```python
# Flag complaints that include a consumer-written narrative
df["has_narrative"] = df["consumer_complaint_narrative"].notna()
```

```python
# Check remaining missing values in each column
df.isna().sum().sort_values(ascending=False)
```

```python
# Create useful fields for monthly trends and response timing
df["received_month"] = df["date_received"].dt.to_period("M").astype(str)

df["days_to_send"] = df["date_sent_to_company"] - df["date_received"]
df["days_to_send"] = df["days_to_send"].dt.total_seconds() / 86400
df["days_to_send"] = df["days_to_send"].round(2)

df["is_timely"] = df["timely_response"].str.lower() == "yes"

df[[
    "date_received",
    "received_month",
    "date_sent_to_company",
    "days_to_send",
    "timely_response",
    "is_timely"
]].head()
```

```python
# Creating date-based and response-time fields for analysis
df["received_month"] = df["date_received"].dt.to_period("M").astype(str)

df["days_to_send"] = df["date_sent_to_company"] - df["date_received"]
df["days_to_send"] = df["days_to_send"].dt.total_seconds() / 86400
df["days_to_send"] = df["days_to_send"].round(2)

df["is_timely"] = df["timely_response"].str.lower() == "yes"

df[[
    "date_received",
    "received_month",
    "date_sent_to_company",
    "days_to_send",
    "timely_response",
    "is_timely"
]].head()
```

```python
# Filling missing optional fields with clear labels
df["sub_issue"] = df["sub_issue"].fillna("Not specified")
df["company_public_response"] = df["company_public_response"].fillna("No public response")
df["tags"] = df["tags"].fillna("No tag")
df["state"] = df["state"].fillna("Unknown")
df["zip_code"] = df["zip_code"].fillna("Unknown")

df.isna().sum().sort_values(ascending=False)
```

```python
# Check a few cleaned fields
df[[
    "product",
    "sub_product",
    "issue",
    "sub_issue",
    "company",
    "state",
    "tags",
    "company_public_response",
    "has_narrative"
]].head()
```

```python
## Removing duplicates and unusable rows

rows_before = len(df)

df["complaint_id"] = pd.to_numeric(df["complaint_id"], errors="coerce")
df = df.dropna(subset=["complaint_id"])
df["complaint_id"] = df["complaint_id"].astype("int64")
df = df.drop_duplicates(subset="complaint_id", keep="first")

rows_after = len(df)

print("Rows before:", rows_before)
print("Rows after:", rows_after)
print("Rows removed:", rows_before - rows_after)
print("Duplicate complaint IDs left:", df["complaint_id"].duplicated().sum())
```

```python
# Removing rows missing the core fields needed for this project
df = df.dropna(subset=[
    "date_received",
    "product",
    "issue",
    "company",
    "date_sent_to_company",
    "company_response_to_consumer",
    "timely_response"
])

# Checking final shape after removing unusable rows
df.shape
```

```python
# Final basic checks after cleaning
print("Rows:", len(df))
print("Columns:", df.shape[1])
print("Date range:", df["date_received"].min(), "to", df["date_received"].max())
print("Products:", df["product"].nunique())
print("Companies:", df["company"].nunique())
print("Issues:", df["issue"].nunique())
print("Timely response rate:", round(df["is_timely"].mean() * 100, 2), "%")
```

## Creating summary tables

####Product summary

```python
# Count complaints by product
product_summary = df.groupby("product")["complaint_id"].count().reset_index()

# Renaming the count column
product_summary = product_summary.rename(columns={"complaint_id": "complaint_count"})

# Sorting from highest to lowest complaints
product_summary = product_summary.sort_values("complaint_count", ascending=False)

product_summary
```

####Company summary

```python
# Count complaints by company
company_summary = df.groupby("company")["complaint_id"].count().reset_index()

# Renaming the count column
company_summary = company_summary.rename(columns={"complaint_id": "complaint_count"})

# Sorting from highest to lowest complaints
company_summary = company_summary.sort_values("complaint_count", ascending=False)

company_summary.head(20)
```

####Issue summary

```python
# Count complaints by issue
issue_summary = df.groupby("issue")["complaint_id"].count().reset_index()

# Renaming the count column
issue_summary = issue_summary.rename(columns={"complaint_id": "complaint_count"})

# Sorting from highest to lowest complaints
issue_summary = issue_summary.sort_values("complaint_count", ascending=False)

issue_summary.head(20)
```

####Monthly summary

```python
# Count complaints by month
monthly_summary = df.groupby("received_month")["complaint_id"].count().reset_index()

# Renaming the count column
monthly_summary = monthly_summary.rename(columns={"complaint_id": "complaint_count"})

# Sorting by month
monthly_summary = monthly_summary.sort_values("received_month")

monthly_summary
```

####Product monthly summary

```python
# Count complaints by month and product
product_monthly_summary = df.groupby(
    ["received_month", "product"]
)["complaint_id"].count().reset_index()

# Renaming the count column
product_monthly_summary = product_monthly_summary.rename(
    columns={"complaint_id": "complaint_count"}
)

product_monthly_summary.head()
```

####Response summary

```python
# Count complaints by company response type
response_summary = df.groupby(
    "company_response_to_consumer"
)["complaint_id"].count().reset_index()

# Renaming the count column
response_summary = response_summary.rename(
    columns={"complaint_id": "complaint_count"}
)

# Sort from highest to lowest
response_summary = response_summary.sort_values("complaint_count", ascending=False)

response_summary
```

####State summary

```python
# Count complaints by state
state_summary = df.groupby("state")["complaint_id"].count().reset_index()

# Renaming the count column
state_summary = state_summary.rename(columns={"complaint_id": "complaint_count"})

# Sort from highest to lowest
state_summary = state_summary.sort_values("complaint_count", ascending=False)

state_summary.head(20)
```
